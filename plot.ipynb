```python
file = open("history.txt","r")
```

```python
texts = file.readlines()
```

```python
dice = []
val_dice = []
for text in texts:
    d,v = text.strip().split()
    dice.append(float(d))
    val_dice.append(float(v))
    print(d)
```

```python
import matplot.pyplot as plt
```

```python
x = range(30)
plt.plot(x,dice,label="dice")
plt.plot(x,val_dice,label="val_dice")
plt.grid()
plt.legend()
plt.show()
```
